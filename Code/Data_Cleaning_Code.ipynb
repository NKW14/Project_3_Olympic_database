```python


df_results_2 = df_results[df_results["event_title"].isin(olympic_events)]
df_results_p_2 = df_results_p[df_results_p["event_title"].isin(olympic_events)]


print(df_results_2)
print(df_results_p_2) 

medals = ['GOLD', "SILVER", "BRONZE"]
df_results_clean = df_results_2[df_results_2["medal_type"].isin(medals)]
df_results_clean
```

```python

def ms_conversion(number_in_ms):
    if ":" in number_in_ms:
        return number_in_ms
    elif "est" in number_in_ms:
        return number_in_ms.replace("est", "").strip()
    else:
        seconds = dt.timedelta(milliseconds=int(number_in_ms)).total_seconds()
        hours = seconds//3600
        minutes = (seconds % 3600)//60
        sec = seconds % 60 
        return "{:02d}:{:02d}:{:.3f}".format(int(hours),int(minutes),round(sec,3))

df_results_clean['value_unit'] = df_results_clean['value_unit'].map(ms_conversion)
df_results_clean
```

```python
df_results_p_2['value_unit'] = df_results_p_2['value_unit'].astype("str")
df_results_p_2["value_unit"] = df_results_p_2['value_unit'].map(ms_conversion)
df_results_p_2
```

```python
df_combined_diff_cols = pd.concat([df_results_p_2, df_results_clean], ignore_index=True)
df_combined_diff_cols.insert(0, "Team_id", range(1001, 1001 + len(df_combined_diff_cols)))
df_combined_diff_cols.insert(0, "Medal_id", range(8001, 8001 + len(df_combined_diff_cols)))
df_combined_diff_cols = df_combined_diff_cols[["Medal_id","Team_id", "event_title","country_name", "value_unit", "medal_type"]]
print(df_combined_diff_cols)

#df_combined_diff_cols.to_csv('Results_new.csv', index=False)
```
